## Grouping the data by sex

```python
Grouped_by_sex = []
for i in range(12):
    Grouped_by_sex.append(DATA_2024[i][['BEN_SEX_COD',
 'FLT_PAI_MNT',
 'FLT_REM_MNT']])
    

dataframes_with_months = [
    df.with_columns(pl.lit(i + 1).alias("month"))  # Month is i+1 since range starts at 0
    for i, df in enumerate(Grouped_by_sex)
]

# Concatenate all DataFrames
df_all = pl.concat(dataframes_with_months)

# Group by month and sex, then sum paid and reimbursed amounts
df_grouped = (
    df_all.group_by(["month", "BEN_SEX_COD"])
    .agg([
        pl.col("FLT_PAI_MNT").sum().alias("Total_Paid"),
        pl.col("FLT_REM_MNT").sum().alias("Total_Reimbursed")
    ])
)

print(df_grouped)


```

```python

# df_grouped.sort_values(by=["month", "BEN_SEX_COD"])

df_grouped["RAC"] = df_grouped["Total_Paid"] - df_grouped["Total_Reimbursed"]

df_grouped.sort_values(by=["month", "BEN_SEX_COD"])
```

```python
data_by_sex = df_grouped[df_grouped["BEN_SEX_COD"].isin([1,2])]

data_by_sex.sort_values(by=["month", "BEN_SEX_COD"], inplace=True)

```

## Grouped the data by age and sex

```python
Grouped_by_age_and_sex = []
for i in range(12):
    Grouped_by_age_and_sex.append(
        DATA_2024[i]
        .select(['BEN_SEX_COD', 'AGE_BEN_SNDS', 'FLT_PAI_MNT', 'FLT_REM_MNT'])
        .group_by('AGE_BEN_SNDS', 'BEN_SEX_COD')
        .sum()
    )
```

## Grouping the data by Age

```python
Grouped_by_age = []
for i in range(12):
    Grouped_by_age.append(DATA_2024[i][['AGE_BEN_SNDS',
 'FLT_PAI_MNT',
 'FLT_REM_MNT']].group_by('AGE_BEN_SNDS').sum())
```

## Grouping the data by Region 


```python
Grouped_by_region = []
for i in range(12):
    Grouped_by_region.append(DATA_2024[i][['BEN_RES_REG',
 'FLT_PAI_MNT',
 'FLT_REM_MNT']].group_by('BEN_RES_REG').sum())
```

## Grouping the data by Nature of insurance

```python
Grouped_by_asu_nat = []
for i in range(12):
    Grouped_by_asu_nat.append(DATA_2024[i][['ASU_NAT',
 'FLT_PAI_MNT',
 'FLT_REM_MNT']].group_by('ASU_NAT').sum())
```

## Sommes des consommations par région pour chaque mois

```python
df_region_merged = pl.concat(Grouped_by_region)
df_region_merged = df_region_merged.group_by("BEN_RES_REG").sum()
df_region_merged = df_region_merged.sort('BEN_RES_REG')

df_region_merged.columns = ["Région", "Frais à payer", "Frais remboursés"]

df_region_merged = df_region_merged.with_columns(
    (pl.col("Frais à payer") - pl.col("Frais remboursés")).alias("Reste à charge")
)



# télécharger en CSV

df_region_merged.to_pandas().to_csv('df_region_merged.csv', index=False)
```

## Somme des dépenses et de remboursement par tranche d'âge et par sexe

```python
df_age_sexe_merged = pl.concat(Grouped_by_age_and_sex)
df_age_sexe_merged = df_age_sexe_merged.group_by("AGE_BEN_SNDS", 'BEN_SEX_COD').sum()
df_age_sexe_merged = df_age_sexe_merged.sort('AGE_BEN_SNDS', 'BEN_SEX_COD')

df_age_sexe_merged.columns = ["Age", "Sexe", "Frais à payer", "Frais remboursés"]

df_age_sexe_merged = df_age_sexe_merged.with_columns(
    (pl.col("Frais à payer") - pl.col("Frais remboursés")).alias("Reste à charge")
)


# télécharger en CSV

df_age_sexe_merged.to_pandas().to_csv('df_age_and_sexe_merged.csv', index=False)
```

## Evolution des frais et des remboursements au cours des mois

```python
months = [int(i) for i in range(1, 13)]
Frais_par_mois = [DATA_2024[i]['FLT_PAI_MNT'].sum() for i in range(12)]
Remboursements_par_mois = [DATA_2024[i]['FLT_REM_MNT'].sum() for i in range(12)]


```

```python
dataframe = pd.DataFrame({'Mois': months, 'Frais à payer': Frais_par_mois, 'Frais remboursés': Remboursements_par_mois})

dataframe['Reste à charge'] = dataframe['Frais à payer'] - dataframe['Frais remboursés']

dataframe.to_csv('df_mois_merged.csv', index=False)
```
